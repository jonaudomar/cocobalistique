```python
class Projectile:
    """
    Projectile soumis uniquement à la gravité.
    """

    def __init__(self, v0, alpha, g=9.81):
        """
        v0    : vitesse initiale (m/s)
        alpha : angle de tir (radians)
        g     : accélération de la pesanteur (m/s²)
        """
        self.v0 = v0
        self.alpha = alpha
        self.g = g

    # ---------- Cinématique ----------

    def position(self, t):
        """
        Position (x, y) en fonction du temps.
        """
        x = self.v0 * np.cos(self.alpha) * t
        y = self.v0 * np.sin(self.alpha) * t - 0.5 * self.g * t**2
        return x, y

    def vitesse(self, t):
        """
        Norme de la vitesse en fonction du temps.
        """
        vx = self.v0 * np.cos(self.alpha)
        vy = self.v0 * np.sin(self.alpha) - self.g * t
        return np.sqrt(vx**2 + vy**2)

    def trajectoire(self, x):
        """
        Équation de la trajectoire Y(X).
        """
        return (
            x * np.tan(self.alpha)
            - (self.g / (2 * self.v0**2 * np.cos(self.alpha)**2)) * x**2
        )

    # ---------- Grandeurs caractéristiques ----------

    def temps_de_vol(self):
        """
        Temps total de vol.
        """
        return 2 * self.v0 * np.sin(self.alpha) / self.g

    def portee(self):
        """
        Portée maximale.
        """
        return (self.v0**2 / self.g) * np.sin(2 * self.alpha)

    def hauteur_max(self):
        """
        Hauteur maximale atteinte.
        """
        return (self.v0**2 * np.sin(self.alpha)**2) / (2 * self.g)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres physiques
v0 = 10.0                     # m/s
alpha = np.deg2rad(45)        # radians
g = 9.81                      # m/s²

# Portée (pour borner l'axe X)
R = (v0**2 / g) * np.sin(2 * alpha)

# Axe horizontal
x = np.linspace(0, R, 50)

# Trajectoire Y(X)
y = x * np.tan(alpha) - (g / (2 * v0**2 * np.cos(alpha)**2)) * x**2

# Tracé
plt.figure()
plt.plot(x, y)
plt.xlabel("X (m)")
plt.ylabel("Y (m)")
plt.title("Trajectoire du projectile : Y en fonction de X")
plt.grid(True)
plt.show()
```

```python
# temps au point d'impact tR à la portée R
tR = 2*v0*np.sin(alpha)/g

t = np.linspace(0, tR, 50)

#norme de la vitesse

v = np.sqrt(v0**2-2*v0*g*t*np.sin(alpha) + (g*t)**2)
```

```python
v
```

```python
plt.figure()
plt.plot(t, v)
plt.xlabel("t (secondes)")
plt.ylabel("Vitesse (m/s)")
plt.title("Vitesse du projectile : V en fonction de t")
plt.grid(True)
plt.show()
```
